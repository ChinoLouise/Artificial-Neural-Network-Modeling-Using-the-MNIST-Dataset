### Preparing the Data for Training ###

##### Flattening the Image Data #####

```python

import numpy as np

x_train_reshaped = np.reshape(x_train, (60000, 784))
x_test_reshaped = np.reshape(x_test, (10000, 784))

print('x_train_reshaped shape: ', x_train_reshaped.shape)
print('x_test_reshaped shape: ', x_test_reshaped.shape)
```

```python
x_train_reshaped[0]
```

```python
set(x_train_reshaped[0])
```

```python
set(x_test_reshaped[0])
```

##### Normalize the Image Data #####

```python

x_train_norm = x_train_reshaped/255
x_test_norm = x_test_reshaped/255
```

```python
set(x_train_norm[0])
```

```python
set(x_test_norm[0])
```

##### Categorized the Label Data #####

```python

import tensorflow.keras as keras
num_categories = 10

y_train_encoded  = keras.utils.to_categorical(y_train, num_categories)
y_test_encoded  = keras.utils.to_categorical(y_test, num_categories)
```

```python
y_train_encoded
```

```python
y_test_encoded
```

##### Validating Shapes After Preprocessing Steps #####

```python

print('x_train_shape: ', x_train.shape)
print('y_train_encoded: ', y_train_encoded.shape)
print('x_test_shape: ', x_test.shape)
print('y_test_encoded: ', y_test_encoded.shape)
```

### Creating a Model ###

##### Instantiating the Model #####

```python

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
	
```

##### Creating the Input Layer #####

```python

model.add(Dense(units = 512, activation = "relu", input_shape = (784, )))
```

##### Creating the Input Layer #####

```python

model.add(Dense(units = 512, activation = "relu", input_shape = (784, )))
```

##### Creating the Hidden Layer #####

```python

model.add(Dense(units = 512, activation = "relu"))
```

##### Creating the Output Layer #####

```python

model.add(Dense(units = 10, activation = "softmax"))
```

##### Summarizing the Model #####

```python

model.summary()
```

##### Compiling the Model #####

```python

model.compile(loss = "categorical_crossentropy", metrics = ["accuracy"])
```

### Training the Model ###

```python

history = model.fit(x_train_norm, y_train_encoded, epochs=50, validation_data = (x_test_norm, y_test_encoded), verbose = 1)
```

##### Monitoring Training and Validation Accuracy During Training #####

```python

plt.plot(history.history["accuracy"])
plt.plot(history.history["val_accuracy"])

plt.xlabel("epoch")
plt.ylabel("accuracy")
```

##### Making a Prediction #####

```python

y_pred_probs = model.predict(x_test_norm)
 
y_pred = np.argmax(y_pred_probs, axis=1)
y_true = np.argmax(y_test_encoded, axis=1)
```

```python

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

print("Accuracy:", accuracy_score(y_true, y_pred))
print("Precision:", precision_score(y_true, y_pred, average='macro'))
print("Recall:", recall_score(y_true, y_pred, average='macro'))
print("F1-score:", f1_score(y_true, y_pred, average='macro'))

print("\nClassification Report:\n")
print(classification_report(y_true, y_pred))
```

### Visualizing Model Predictions on Test Data ###

```python

plt.figure(figsize=(20, 12))

num_images = 30
random_indices = np.random.choice(len(x_test), num_images, replace=False)

for i, idx in enumerate(random_indices):
    plt.subplot(5, 6, i+1)
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    
    pred = np.argmax(y_pred_probs[idx])
    gt = y_test[idx]
    
    col = 'b' if pred == gt else 'r'

    plt.xlabel('prediction={} \n correct output={}'.format(pred, gt), color=col, fontsize=12)
    
    plt.imshow(x_test[idx], cmap='binary')

plt.tight_layout()
plt.show()
```
